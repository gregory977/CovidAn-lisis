```python
##Revisar de nuevo archivo csv, caracteres y delimitaci칩n y columnas

import pandas as pd

ruta_archivo = '/content/Casos_positivos_de_COVID-19-Cund-Boy.csv'

df = pd.read_csv(ruta_archivo, delimiter=';')
print(df.columns)
```

```python
# Crear tablas para el modelo de datos##
municipality = df[['id_case', 'id_department', 'name_municipality', 'id_municipality']]

department = df[['id_case', 'id_department']]

gender = df[['id_case', 'gender']]

type_contagion = df[['id_case', 'type_contagion']]

estatus = df[['id_case', 'status']]

##Descargas para usar en Power BI
municipality.to_csv('/content/hoja_municipality.csv', index=False)
department.to_csv('/content/hoja_department.csv', index=False)
gender.to_csv('/content/hoja_gender.csv', index=False)
type_contagion.to_csv('/content/hoja_type_contagion.csv', index=False)
estatus.to_csv('/content/hoja_status.csv', index=False)
```
